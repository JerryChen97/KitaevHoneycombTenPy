```python
Jx = 1.0
Jy = 1.0
Jz = 1.0

S = 1.5

Lx = 6
Ly = 4
shift = 0
bc = 'periodic'
bc_MPS = 'infinite'

chi_list = [int(2**i) for i in range(2, 9)]

result_dict = {}

initial_psi = None
for chi in chi_list:
    param = (Jx, Jy, Jz, chi)
    result = run_atomic(
        Jx=Jx, Jy=Jy, Jz=Jz,
        Lx=Lx, Ly=Ly, 
        S=S,
        shift=shift,
        chi=chi,
        bc=bc, bc_MPS=bc_MPS,
        initial_psi=initial_psi,
    )
    initial_psi = result['psi'].copy()
    result_dict[param] = result
```
