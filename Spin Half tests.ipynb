```python
from KitaevHoneycomb import run_atomic
```

```python
import numpy as np
```

```python
Jx = 1.0
Jy = 1.0
Jz = 1.0

Lx = 6
Ly = 4
shift = 0
bc = 'periodic'
bc_MPS = 'infinite'

chi_list = [2**i for i in range(1, 9)]

result_dict = {}

for J in np.linspace(0, 1.0, 11):
    J = J.round(decimals=2)
    initial_psi = None
    for chi in chi_list:
        param = (J, J, Jz, chi)
        if param not in result_dict:
            result = run_atomic(
                Jx=J, 
                Jy=J, 
                Jz=Jz, 
                Lx=Lx, Ly=Ly, 
                chi=chi,
                shift=shift,
                bc=bc, bc_MPS=bc_MPS,
                initial_psi=initial_psi,
            )
        else:
            result = result_dict[param]
        initial_psi = result['psi'].copy()
        result_dict[param] = result
```
